```
ig_char_p = 'gdrive/My Drive/aspect_ai/ignore_char.txt'
ig_unk_p = 'gdrive/My Drive/aspect_ai/ignore_unk.txt'
with open(ig_char_p) as f:
    ignore_char = json.load(f)
with open(ig_unk_p) as f:
    ignore_unk = json.load(f)
```

```
p = re.compile(r'^[\d]+[A-Za-z]*')
p2 = re.compile(r'^[A-Za-z]+[\d]+')
def token(row):
    content = row['content'].strip('"').strip()
    content = content.replace('★★★★★', '非常好')
    content = content.replace('★★★★☆', '好')
    content = content.replace('★★★☆☆', '一般')
    content = content.replace('★★☆☆☆', '差')
    content = content.replace('★☆☆☆☆', '很差')
    
    content = content.replace('★★★★', '好')
    content = content.replace('★★★', '一般')
    content = content.replace('★★', '差')
    
    content = content.replace('⭐️⭐️⭐️⭐️⭐️', '非常好')
    content = content.replace('⭐️⭐️⭐️⭐️', '好')
    content = content.replace('⭐️⭐️⭐️', '一般')
    content = content.replace('⭐️⭐️', '差')
    content = content.lower()
    content = HanziConv.toSimplified(content) 
    content_token = list(jieba.cut(content))
    tokens = []
    for x in content_token:
        if x in ignore_unk:
            continue
        if x in ignore_char:
            continue
        if p.findall(x) or p2.findall(x):
            continue
        if x in stopwords:
            continue
        x = x.strip('\n').strip(' ')
        if x == '':
            continue
        tokens.append(x)
    
    row['content_token_clean'] = ' '.join(tokens)
#     if row['location_traffic_convenience'] in [0, 1, -2, -1]:
#         label = row[list(range(2, 22))].values.tolist()
#         oh = encode_one_hot(4, label)
#         row['y'] = oh.flatten().astype(np.int32).tolist()
    return row
```

```
def pre(row):
  content_token = row['content_token']
#   content_token = content_token.lower()
#   content_token = HanziConv.toSimplified(content_token)
  
  words = content_token.split(' ')
  temp = []
  for w in words:
    if p.findall(w) or p2.findall(w):
      print(w)
      w = 'unk'
    w = w.strip('\n').strip(' ')
    if w == '':
#       print('empty char')
      continue
    temp.append(w)
  
  row['content_token'] = ' '.join(temp)
  
  return row
```

```
import numpy as np
import pandas as pd
```

```
def encode_one_hot(class_num, label):
    """
    one - hot 
    对角矩阵的方式实现，不论类别出现的顺序怎样都会得到一样的结果，
    用pd的get_dummies会有类别出现顺序不一样得到的结果也不一样        
    :param class_num: 
    :param label: 为数字list，比如 [0,1,2]
    :return: np.array
    """
    met = np.eye(class_num)
    return met[label]
```

```
def process(row):
    label = eval(row['label'])
    oh = encode_one_hot(4, label)
    row['y'] = oh.flatten().astype(np.int32).tolist()

    return row
```
